Add a column “Range” to the existing dataframe for below condition
a. 1 for Effective_Price > 500
b. 0.5 otherwise

```python
def check_efc_price(effectivePrice):
  if effectivePrice > 500.0:
    return 1
  else:
    return 0.5

df['Range'] = df['Effective_Price'].apply(lambda x:check_efc_price(x))
```
